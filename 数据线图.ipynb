```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 
#data = pd.read_csv("your_data.csv")  # 
file_path = 'C:/Users/rainc/Desktop/1-1-111.txt'
data = pd.read_csv(file_path, sep="\t")
#  x  y
x_col = "J (mA/cm2)"
y_cols = ["EQE(%)", "Cur Eff (cd/A)", "L (cd/m2)"]

#  long-form 
data_long = data[[x_col] + y_cols].melt(id_vars=x_col, 
                                        value_vars=y_cols,
                                        var_name="Metric", 
                                        value_name="Value")

# 
sns.set_theme(style="darkgrid")
sns.lineplot(data=data_long, x=x_col, y="Value", hue="Metric", linewidth=1)

plt.title("Device Performance vs Current Density")
plt.xlabel(x_col)
plt.ylabel("Value")
plt.legend(title="Metric")
plt.show()
```
