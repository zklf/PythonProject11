```python
import numpy as np
import pandas as pd

# 每隔10度，从0到90度
angles_deg = np.arange(0, 91, 10)
angles_rad = np.deg2rad(angles_deg)  # 转换为弧度

# 计算朗伯发射分布（归一化）
intensities = np.cos(angles_rad)

# 保存到DataFrame
df = pd.DataFrame({
    "Angle (°)": angles_deg,
    "Relative Intensity": intensities
})

# 输出到CSV文件
df.to_csv("lambertian_emission.csv", index=False)

# 可视化（可选）
#import matplotlib.pyplot as plt
#plt.plot(angles_deg, intensities, marker='o')
#plt.xlabel("Angle (°)")
#plt.ylabel("Relative Intensity (cosθ)")
#plt.title("Lambertian Emission (0° = 1)")
#plt.grid(True)
#plt.show()
```
